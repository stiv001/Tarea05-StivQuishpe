```python
from scipy.optimize import newton
import sympy as sp
import numpy as np
``
f = lambda x: -(x**3) - np.cos(x)
df = lambda x: -3*x**2 + np.sin(x)
p0 = -1
tol = 1.48e-08

try:
    raiz = newton(f, p0, fprime=df, tol=tol)
    print(f"p2 = {raiz}")
except RuntimeError as e:
    print("El método fracasó al no converger a una raíz.")
```
